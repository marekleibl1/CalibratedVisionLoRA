```python
from matplotlib import pyplot as plt 
import numpy as np
import json

plt.style.use('dark_background')


def plot_results(json_path, title, xlabel, ylabel): 
    
    with open(json_path, 'r') as f: 
        d = json.load(f)
    
    data = [ [int(rank), loss] for rank, losses in d.items() for loss in losses]
    data = list(zip(*data))
    plt.scatter(data[0], data[1], c='blue', alpha=.8, marker='x')
    
    data = [ [int(rank), np.mean(losses)] for rank, losses in d.items()]
    data = list(zip(*data))
    plt.scatter(data[0], data[1], c='red', marker='x')
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(np.arange(min(data[0]), max(data[0])+1, 1)) # Ensure x-axis uses integer values
    plt.show()


json_path = '../results/lora_ranks.json'
plot_results(json_path, 
            'Validation Losses with Different LoRA Ranks', 
            'LoRA rank', 'Validation loss')

```
